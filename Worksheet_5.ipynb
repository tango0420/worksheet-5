```python
import numpy as np

# Features (X) and Labels (Y) from the dataset
X = dataset[['Math', 'Reading']].values
Y = dataset['Writing'].values

X_transposed = X.T

W = np.zeros((X_transposed.shape[0], 1))

Y = Y.reshape(-1, 1)

print("Shape of X_transposed (Feature Matrix):", X_transposed.shape)
print("Shape of W (Weight Matrix):", W.shape)
print("Shape of Y (Target Matrix):", Y.shape)
```

```python
# Calculate predictions
Y_pred = np.dot(W.T, X_transposed)

# Display shapes
print("Shape of W^T (Transposed Weight Matrix):", W.T.shape)
print("Shape of X (Feature Matrix):", X_transposed.shape)
print("Shape of Y_pred (Predicted Targets):", Y_pred.shape)

# Print the first few predictions
print("\nPredicted Y (first 5 values):")
print(Y_pred[:, :5])  # First 5 predictions
```

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Display the shapes of the resulting sets
print("Training Feature Matrix (X_train):", X_train.shape)
print("Test Feature Matrix (X_test):", X_test.shape)
print("Training Target Matrix (Y_train):", Y_train.shape)
print("Test Target Matrix (Y_test):", Y_test.shape)
```
